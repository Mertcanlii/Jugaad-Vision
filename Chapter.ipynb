```python
from __future__ import print_function
import argparse
import cv2
from google.colab.patches import cv2_imshow


image = cv2.imread("2022-11-23.png")
cv2_imshow(image) #resimi göstermek için bu kod
(b, g, r) = image[0, 0] """On Line 12, we grab the pixel located at (0, 0) – the top left corner of the image. 
This pixel is represented as a tuple.
Again, OpenCV stores RGB pixels in reverse order, so when
we unpack and access each element in the tuple, we are actually
viewing them in BGR order. """
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r,
                                                              g, b))
image[0, 0] = (0, 0, 255) #we paint the pixel to the red.
(b, g, r) = image[0, 0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r,
                                                              g, b))
corner = image[0:100, 0:100]
cv2_imshow(corner)


image[0:100, 0:100] = (0, 255, 0)

cv2_imshow(image)
cv2.waitKey(0)
```

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

canvas = np.zeros((300, 300, 3), dtype = "uint8")
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)

cv2.waitKey(0)


red = (0, 0, 255)
cv2.line(canvas, (300, 0), (0, 300), red, 3)

cv2.waitKey(0)
cv2.rectangle(canvas, (10, 10), (60, 60), green)


cv2.waitKey(0)

cv2.rectangle(canvas, (50, 200), (200, 225), red, 5)

cv2.waitKey(0)

blue = (255, 0, 0)
cv2.rectangle(canvas, (200, 50), (225, 125), blue, -1)
cv2_imshow(canvas)
cv2.waitKey(0)


canvas = np.zeros((300, 300, 3), dtype = "uint8")
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
white = (255, 255, 255)

for r in range(0, 175, 25):
  cv2.circle(canvas, (centerX, centerY), r, white)
cv2_imshow(canvas)
cv2.waitKey(0)

for r in range(0, 175, 25):
  cv2.circle(canvas, (centerX, centerY), r, white)

cv2_imshow(canvas)
cv2.waitKey(0)
for i in range(0, 25):
  radius = np.random.randint(5, high = 200)
  color = np.random.randint(0, high = 256, size = (3,)).tolist()
  pt = np.random.randint(0, high = 300, size = (2,))
  cv2.circle(canvas, tuple(pt), radius, color, -1)

cv2_imshow(canvas)
cv2.waitKey(0)
```
